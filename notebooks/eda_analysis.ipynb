```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("../data/airbnb_listings_london.csv")

# Preview dataset
display(data.head())

# Basic info
print("Dataset Info:")
display(data.info())

# Missing values per column
print("Missing Values:")
display(data.isnull().sum().sort_values(ascending=False))

# Summary statistics
print("Summary Statistics:")
display(data.describe(include='all'))

# Unique values in each column (for categoricals)
categorical_cols = data.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\nUnique values in '{col}':")
    print(data[col].unique())

# Distribution of room types
plt.figure(figsize=(8,4))
data['room_type'].value_counts().plot(kind='bar')
plt.title("Distribution of Room Types")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

# Check if neighbourhood is usable as a shard key
plt.figure(figsize=(10,5))
data['neighbourhood'].value_counts().head(20).plot(kind='bar')
plt.title("Top 20 Neighbourhoods by Listings")
plt.ylabel("Number of Listings")
plt.xticks(rotation=90)
plt.show()

# If 'neighbourhood_group' is empty, confirm it:
if data['neighbourhood_group'].isnull().all():
    print("Column 'neighbourhood_group' is completely empty.")
```
